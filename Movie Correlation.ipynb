```python
# Importing libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Reading the data
df = pd.read_csv(r'C:\Users\User\Downloads\movies.csv')
```

```python
# Replacing 'Null' values with the mean for 'gross' and 'budget' columns

# Get list of float cols with missing values
num_cols=df.select_dtypes(include=['float']).columns

# Filling missing values with the mean values of the cols
df[num_cols]= df[num_cols].fillna(df[num_cols].mean())
```

```python
# Count missing values in each column

print(df.isnull().sum())
```

```python
# Count missing values in budget column

print(df['budget'].isnull().sum())
```

```python
# Deleting rows that contain null values (dtype object)

df = df.dropna()
```

```python
# Count missing values in each column

print(df.isnull().sum())
```

```python
#sort gross values in descending order
df=df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#looking at correlation
df.corr(method='pearson')
```

```python
correlation_matrix=df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
#changing data type of object cols to category data type
df_numerized= df
for col_name in df_numerized.columns:
    if(df_numerized[col_name]).dtype=='object':
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
correlation_matrix=df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat= df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs= corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs > 0.5) & (sorted_pairs != 1) & ~(sorted_pairs.duplicated())]
high_corr
```
